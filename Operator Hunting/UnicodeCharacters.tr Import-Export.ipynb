```python
import pandas as pd
import numpy as np
import unicodedata
```

```python
classes = ['WhiteSp', 'Letter', 'Alias', 'NonBreakingBeforeLetter', 'Prefix',
       'Infix', 'LargeOp:Prefix', 'CompoundPrefix', 'InfixOpen', 'Close',
       'Open', 'Postfix', 'Auto', 'NonBreakingAfterLetter']
for c in classes:
    print(c)
```

```python
ops=df[df['Class / Affix'].isin(['Prefix',
       'Infix', 'LargeOp:Prefix', 'CompoundPrefix', 'InfixOpen', 'Close',
       'Open', 'Postfix'])]
```

```python
outname = 'operators.csv'
ops.to_csv(outname)
```

```python
def hex_string_to_int(hex):
    return int(hex, 0)
```

```python
ops['Unicode Number'] = ops['Character Code'].apply(hex_string_to_int)
```

```python
ops['Unicode Character'] = ops['Unicode Number'].apply(chr)
```

```python
def getUnicodeName(c):
    try:
        return unicodedata.name(c)
    except ValueError as e:
        pass
    return None
    
```

```python
ops['Unicode Name'] = ops['Unicode Character'].apply(getUnicodeName)
```

```python
def phraseToCamel(phrase):
    return ''.join([s.capitalize() for s in phrase.split()])
```

```python
def fix_names(row):
    if row['Unicode Name'] == 'LEFT ANGLE BRACKET':
        return 'FullwidthLeftAngleBracket'
    if row['Unicode Name'] == 'RIGHT ANGLE BRACKET':
        return 'FullwidthRightAngleBracket'
    elif (row['Long Name']=="\[]") & (row['Unicode Name'] is not None):
        return phraseToCamel(row['Unicode Name'])
    elif (row['Long Name']=="\[]") & (row['Unicode Name'] is None):
        return None

    return row['Long Name'][2:-1]
```

```python
ops['Long Name'] = ops.apply(fix_names, axis=1)
```

```python
ops.to_csv('operators.csv')
```
