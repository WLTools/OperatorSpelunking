```python
import pandas as pd

import numpy as np
```

```python
filename = 'OperatorTable.csv'
ops = pd.read_csv(filename)
# Delete empty rows at the end of the file.
ops = ops[ops['Name'].notnull()]
```

```python
# Reverse the list.
ops = ops.reindex(index=ops.index[::-1])
```

```python
levels = []
previous = 0
newprec = 0
current = 0
counter = 0
for i, row in ops.iterrows():
#     counter += 1
#     if counter >10:
#         break
        
    # Get the next number
    if pd.isna(row['Me']):
        current = row['RightPrec']
    else:
        current = row['Me']
        
    # Compare it to the previous number
    if current == previous:
        levels.append(newprec)
    else:
        # Make a new level.
        previous = current
        newprec += 10
        levels.append(newprec)
    
```

```python
ops['NewPrec'] = pd.Series(levels).values
```

```python
# Reverse back to original.
ops = ops.reindex(index=ops.index[::-1])
```

```python
ops.to_csv(filename)
```
